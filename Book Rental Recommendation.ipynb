```python
df_feature = df_rating[['isbn','user_id']]
df_feature
```

```python
df_rate = df_rating[['rating']]
df_rate
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train, feature_test, rating_train, rating_test = train_test_split (df_feature, df_rate, random_state=1)
print(feature_train.shape)
print(feature_test.shape)
print(rating_train.shape)
print(rating_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(feature_train,rating_train)
feature_predict = linreg.predict(feature_test)
feature_predict
```

```python
from sklearn import metrics
import numpy as np
```

```python
np.sqrt(metrics.mean_squared_error(rating_test,feature_predict))
```
